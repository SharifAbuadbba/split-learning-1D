```
import pandas as pd
import numpy as np

import os
```

```
from google.colab import drive

drive.mount('/content/gdrive')
root_path = 'gdrive/My Drive/Colab Notebooks'
```

```
mean = []
dev = []

for i in range(2, 9):
    print(i)
    df = pd.read_csv(os.path.join(root_path, 'csv', 'accs_conv' + str(i) + '.csv'))
    arr = df['best_test_acc'].values
    print(np.argsort(arr)[::-1])
    print(arr)
    mean.append(np.mean(arr))
    dev.append(np.std(arr))

print(mean)
print(dev)
```

```
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
_, caps, _ = ax.errorbar(np.array([i for i in range(2, 9)]), np.array(mean), yerr=np.array(dev), fmt='-o', capsize=6, color='purple')
ax.set_xlabel('Number of Split Conv. Layers', size=14)
ax.set_ylabel('Accuracy', size=14)
ax.grid(alpha=0.5)
ax.tick_params(labelsize=14)
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
plt.savefig(os.path.join(root_path, 'img', 'accuracy_adding_layers.pdf'), bbox_inches='tight')
```
